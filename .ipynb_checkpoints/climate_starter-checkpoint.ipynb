# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, distinct, text, desc
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# reflect an existing database into a new model

# reflect the tables
```

```python
# We can view all of the classes that automap found
# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
# Save references to each table
# Assign the dow class to a variable called `Dow`
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
first_row = session.query(Measurement).first()
first_row.__dict__
```

```python
first_row = session.query(Station).first()
first_row.__dict__
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
for row in session.query(Measurement.date).all():
    print(row)
```

```python
date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
date = list(np.ravel(date))
date[0]
```

```python
#latest date in table
last_date = dt.datetime.strptime(date[0],'%Y-%m-%d' )
```

```python
# Calculate the date 1 year ago from the last data point in the database

begin_date = last_date - timedelta(days=365)
begin_date
```

```python
# Perform a query to retrieve the data and precipitation scores for the past year
annual_rainfall = session.query(Measurement.station,Measurement.date,  Measurement.prcp).filter(Measurement.date >= begin_date).filter(Measurement.date <= last_date).order_by(Measurement.station).order_by(Measurement.date).all()
annual_rainfall
```

```python
rainfall_df = pd.DataFrame(annual_rainfall)
rainfall_df['station'].unique()
```

```python

rainfall_new_df = rainfall_df
rainfall_df = rainfall_df.pivot(index = 'date', columns = 'station')
rainfall_df
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
summary_statistics_df = pd.DataFrame()
for station in station_list:
    summary = rainfall_df.loc[:,('prcp', station)].describe()
    var = pd.Series(rainfall_df.loc[:,('prcp', station)].var(), index = ['var'])
    sem = pd.Series(rainfall_df.loc[:,('prcp', station)].sem(), index = ['sem'])
    summary = pd.Series(summary.append(to_append = (var,sem)),name = station)
    summary_statistics_df = pd.concat([summary_statistics_df,summary], axis = 1)
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
station_ranking = session.query(Measurement.station, func.count(Measurement.date).label('station_count')).group_by(Measurement.station).order_by(desc('station_count')).all()
station_ranking
```

```python
most_active_station = station_ranking[0][0]
most_active_station
```

```python
#keep this as an example of how to do an equality
list(np.ravel(session.query(Measurement.station).first()))[0]=='USC00519397'
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
sel = [Measurement.station, func.min(Measurement.tobs),\
      func.max(Measurement.tobs),func.avg(Measurement.tobs)]

session.query(*sel).filter_by(station = most_active_station).all()
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
temp_count_station = session.query(Measurement.station, func.count(Measurement.tobs).label('temperature_count')).group_by(Measurement.station).order_by(desc('temperature_count')).all()
temp_count_station
```

```python
station_most_temp = temp_count_station[0][0]
station_most_temp
```

```python
temp_past_year = session.query(Measurement.station,Measurement.date,  Measurement.tobs).filter(Measurement.date >= begin_date).filter(Measurement.date <= last_date).filter_by(station = station_most_temp).order_by(Measurement.station).order_by(Measurement.date).all()
temp_past_year
```

```python
only_temp_list = [temp[2] for temp in temp_past_year]
only_temp_list
```

```python
bins = 12
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(only_temp_list, bins = bins)
ax.set_xlabel('temperature')
ax.set_ylabel('count')
ax.set_title('histogram of temperature measurements')
ax.legend(labels = 'tobs')
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
