# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, distinct, text, desc
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# reflect an existing database into a new model

# reflect the tables
```

```python
# We can view all of the classes that automap found
# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
# Save references to each table
# Assign the dow class to a variable called `Dow`
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
first_row = session.query(Measurement).first()
first_row.__dict__
```

```python
first_row = session.query(Station).first()
first_row.__dict__
```

# Temperature Analysis 1

```python
date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
date = list(np.ravel(date))
date[0]
```

```python
#latest date in table
last_date = dt.datetime.strptime(date[0],'%Y-%m-%d' )
```

```python
# Calculate the date 1 year ago from the last data point in the database

begin_date = last_date - timedelta(days=365)
begin_date
```

```python
avg_temperature_june_dec = session.query(func.strftime("%m", Measurement.date),func.avg(Measurement.tobs)).order_by(Measurement.station).filter((func.strftime("%m", Measurement.date) == "06") | (func.strftime("%m", Measurement.date) == "12")).group_by(func.strftime("%m", Measurement.date)).all()
avg_temperature_june_dec
```

```python
# Perform a query to retrieve the temperature data for the entire set
avg_temperature_june_dec = session.query( func.avg(Measurement.tobs)).order_by(Measurement.station).filter((func.strftime("%m", Measurement.date) == "06") | (func.strftime("%m", Measurement.date) == "12")).group_by(func.strftime("%m", Measurement.date)).all()
avg_temperature_june_dec
```

```python
temperature_june = session.query(Measurement.tobs).order_by(Measurement.station).filter((func.strftime("%m", Measurement.date) == "06")).all()
temperature_june = list(np.ravel(temperature_june))
temperature_june
```

```python
np.mean(temperature_june)
```

```python
temperature_dec = session.query( Measurement.tobs).order_by(Measurement.station).filter((func.strftime("%m", Measurement.date) == "12")).all()
temperature_dec = list(np.ravel(temperature_dec))
temperature_dec
```

```python
np.mean(temperature_dec)
```

```python
sts.ttest_ind(temperature_june, temperature_dec)
```

We must perform an unpaired t-test because the data sizes are different.  There are more temperatures in June than December.  The hypothesis is that the 2 temperatures are different.  The null hypothesis is that they are the same.  Since the p-value is essentially 0 the null hypothesis is rejected and the temperature differences are said to be different.
